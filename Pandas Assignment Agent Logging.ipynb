```python
#1. Converting dates into datetime type

agent_perf['Date'] = pd.to_datetime(agent_perf['Date'])
```

```python
#2. new column with weekly datetime

agent_perf['week'] = agent_perf['Date'].dt.isocalendar().week
```

```python
#3. sorting agents by considering only ones who have logged in
agent_p = agent_perf[agent_perf['Total Chats']>0]
```

```python
#q1 - Find out their avarage rating on weekly basis keep this in a mind that they have 2 leaves.
```

```python
#q9 - average weekly response time for each agent 
```

```python
#Average Response time and Average Resolution Time is in object type
# changing it to float/int

resp = agent_p['Average Response Time'].str.split(":")
reso = agent_p['Average Resolution Time'].str.split(":")

agent_p['Conv Average Response Time'] = (resp.str[0].astype(int) * 3600 + resp.str[1].astype(int) * 60 + resp.str[2].astype(int))
agent_p['Conv Average Resolution Time'] = (reso.str[0].astype(int) * 3600 + reso.str[1].astype(int) * 60 + reso.str[2].astype(int))
```

```python
# calculating avg weekly response time
#in this table we will again take data of agents who have logged in only 'agent_p', as if we take all data _perf then avg will change!

avg_weekly_resp_time = agent_p.groupby(['week','Agent Name'])['Conv Average Response Time'].mean()/60
avg_weekly_resp_time = avg_weekly_resp_time.to_frame(name = 'Avg Weekly Resp. time in min')
avg_weekly_resp_time
```

```python
#converting 'duration' type into int, and converting into seconds

duration = agent_logging['Duration'].str.split(":")
agent_logging['Duration in Hours'] = (duration.str[0].astype(int)*3600 + duration.str[1].astype(int)*60 + duration.str[2].astype(int))/3600
agent_logging['Date'] = pd.to_datetime(agent_logging['Date'])
agent_logging['Week'] = agent_logging['Date'].dt.isocalendar().week
```

```python
agent_logging
```

```python
weekly_total_hr = agent_logging.groupby(['Agent','Week'])['Duration in Hours'].sum()
weekly_total_hr = weekly_total_hr.to_frame(name='Hours per week')
weekly_total_hr
```

```python
#q14 - total percentage of active hour for a month
```

```python
total_monthly_hours = (agent_logging.groupby('Agent')['Duration in Hours'].sum()) / (24*30) * 100
total_monthly_hours = total_monthly_hours.to_frame(name = 'Monthly active hours %')
total_monthly_hours
```
