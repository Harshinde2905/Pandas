```python
# 8) Create a new column and try to find out difference between order date and shipment date
```

```python
usa['Conv_OrderDate'] = pd.to_datetime(usa['Order Date'])
usa['Conv_ShipDate'] = pd.to_datetime(usa['Ship Date'])

#usa['date'] = pd.to_datetime(usa['Ship Date'], format='%d-%m-%y')
```

```python
usa['Shipment Duration'] = usa['Conv_ShipDate'] - usa['Conv_OrderDate']
```

```python
# 9) for que 8 find out where duration is more than 10 days.
```

```python
# we have to first convert conv_date column into str format and then strip the 'days' then convert back to int to use the int data to calculate

usa['Shipment Days'] = usa['Shipment Duration'].astype(str)
#usa['Shipment Days'] = usa['Shipment Duration1'].str.replace('days', '')
usa['Shipment Days'] = usa['Shipment Duration1'].astype(int)
```

```python
# 11) group by region and find out the most profitable region.
```
