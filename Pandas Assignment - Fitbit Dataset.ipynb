```python
#q9. how many person are not a regular person with respect to activity try to find out those 
```

```python
health = fitbit.groupby(['Id'])[['Healthy_Day','Calories']].sum()
health.head()
```

```python
health['Logged_days'] = fitbit.groupby('Id')['ActivityDate'].count().sort_values(ascending=False)
health['AvgCalories'] = health['Calories']/health['Logged_days']
health['Health Index'] = health['Healthy_Day'] / health['Logged_days']
```

```python
health.sort_values(by= ['AvgCalories','Healthy_Day'],ascending=False).head()
```

```python
#q9. how many person are not a regular person with respect to activity try to find out those 
```

```python
#1 lets assume to be regular you must walk 5000 steps everyday
```

```python
total_steps = fitbit.groupby('Id')['TotalSteps'].sum()
```

```python
total_steps = total_steps.to_frame(name='total_steps')
total_steps
```

```python
totdays = fitbit.groupby('Id')['ActivityDate'].count()
```

```python
totdays = totdays.to_frame(name='total_days')
```

```python
avg_steps = total_steps['total_steps']/totdays['total_days']
avg_steps = avg_steps.to_frame(name='avg_steps')
```

```python
avg_steps 
```

```python
def steps(s):
    if s > 5000:
        return 'Regular'
    else:
        return 'Not Regular'
```

```python
avg_steps['flag_regular'] = avg_steps['avg_steps'].apply(steps)
```
